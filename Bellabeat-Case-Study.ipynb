```python
activity['ActivityDate'] = pd.to_datetime(activity['ActivityDate'])
sleep['SleepDay'] = pd.to_datetime(
    sleep['SleepDay'],
    format='%m/%d/%Y %I:%M:%S %p'
)
```

```python
sleep = sleep.rename(columns={'SleepDay': 'ActivityDate'})
```

```python
merged_data = pd.merge(
    activity,
    sleep,
    on=['Id', 'ActivityDate'],
    how='inner'
)
```

```python
merged_data[['TotalMinutesAsleep', 'TotalSteps']].corr()
```

```python
### Analysis 3: Sleep Duration vs Daily Steps
plt.figure(figsize=(8,5))
plt.scatter(
    merged_data['TotalMinutesAsleep'],
    merged_data['TotalSteps']
)
plt.title("Sleep Duration vs Daily Steps")
plt.xlabel("Minutes Asleep")
plt.ylabel("Total Steps")
plt.grid(True)
plt.show()
```

```python
## Relationship Between Sleep Duration and Daily Steps

### Business Question

# Does sleep duration affect daily physical activity?

### Statistical Analysis

# The correlation coefficient between **TotalMinutesAsleep** and **TotalSteps** is **-0.19**, 
# indicating a very weak negative correlation.

### Observation

# - Most users sleep between **300 and 550 minutes** (approximately **5–9 hours**) per night.
# - The scatter plot does not show a clear linear relationship between sleep duration and daily steps.
# - A few outliers are present, representing users with unusually short or long sleep durations.

### Business Insight

# Based on this dataset, sleep duration alone is not a reliable indicator of daily physical activity. 
#    Bellabeat should consider combining sleep data with other health metrics to better understand user behavior.
```

```python
### Analysis 4: User Activity Segmentation
def classify_steps(steps):
    if steps < 5000:
        return "Sedentary"
    elif steps < 7500:
        return "Low Active"
    elif steps < 10000:
        return "Somewhat Active"
    else:
        return "Active"
activity['ActivityLevel'] = activity['TotalSteps'].apply(classify_steps)
```

```python
activity['ActivityLevel'].value_counts()
```

```python
activity['ActivityLevel'].value_counts(normalize=True) * 100
```

```python
avg_steps = activity.groupby('Id')['TotalSteps'].mean()
def classify_steps(steps):
    if steps < 5000:
        return "Sedentary"
    elif steps < 7500:
        return "Low Active"
    elif steps < 10000:
        return "Somewhat Active"
    else:
        return "Active"

user_activity = avg_steps.reset_index()
user_activity['ActivityLevel'] = user_activity['TotalSteps'].apply(classify_steps)
# user_activity
```

```python
user_activity['ActivityLevel'].value_counts()
```

```python
user_activity['ActivityLevel'].value_counts(normalize=True) * 100
```

```python
import matplotlib.pyplot as plt

user_activity['ActivityLevel'].value_counts().plot(
    kind='bar',
    figsize=(7,5)
)
plt.title("User Activity Levels")
plt.xlabel("Activity Level")
plt.ylabel("Number of Users")
plt.show()
```

```python
# User Activity Segmentation
## Business Question
# How are users distributed based on their average daily step counts?
## Statistical Analysis
# Users were classified into four activity levels based on their average daily steps:
# - Sedentary (< 5,000 steps)
# - Low Active (5,000–7,499 steps)
# - Somewhat Active (7,500–9,999 steps)
# - Active (10,000+ steps)
## Observation
# - 9 users (27.3%) were classified as **Low Active**.
# - 9 users (27.3%) were classified as **Somewhat Active**.
# - 8 users (24.2%) were classified as **Sedentary**.
# - 7 users (21.2%) were classified as **Active**.
# Most users fall into the **Low Active** and **Somewhat Active** categories, while the **Active** group is the smallest.
## Business Insight
# Bellabeat can improve user engagement by creating personalized activity plans that help users gradually move 
#    from lower activity levels to higher ones.
# Bellabeat has an opportunity to increase user engagement by helping users move from lower activity levels to higher 
# ones through personalized goals, reminders, and motivational challenges.
```

```python
### Analysis 5: Average Sleep Duration by Activity Level
avg_sleep = sleep.groupby('Id')['TotalMinutesAsleep'].mean().reset_index()
user_summary = pd.merge(
    user_activity,
    avg_sleep,
    on='Id'
)
user_summary.groupby('ActivityLevel')['TotalMinutesAsleep'].mean()
```

```python
sleep_by_activity = user_summary.groupby('ActivityLevel')['TotalMinutesAsleep'].mean()
sleep_by_activity.plot(kind='bar', figsize=(7,5))
plt.title("Average Sleep Duration by Activity Level")
plt.xlabel("Activity Level")
plt.ylabel("Average Minutes Asleep")
plt.show()
```

```python
## Observation
# - The highest average sleep duration was observed in the **Low Active** group.
# - The **Active** group had the lowest average sleep duration.
# - Sleep duration differs across activity levels.
# - However, these results should be interpreted carefully because some activity groups contain only a small number of users.
## Business Insight
#The analysis suggests that higher activity levels do not necessarily correspond to longer sleep duration.
#Bellabeat should consider additional factors such as exercise intensity, age, work schedule, 
#stress, and lifestyle when evaluating the relationship between physical activity and sleep.
## Limitation
#The sleep dataset includes only 24 users, while the activity dataset contains 33 users. Therefore, 
#the comparison between activity level and sleep duration is based on a limited sample size, 
#which may affect the generalizability of the findings.
```
